```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction .text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/Courses.csv')
```

```python
df.head(3)
```

We are doing Content Based Filtering using Memory Based Approach.

```python
features = ['CourseTitle', 'Description']
```

Preprocessing for null values

```python
for feature in features:
  df[feature] = df[feature].fillna('')
```

Concatenating two strings

```python
def combined_features(row):
  return row['CourseTitle']+ " "+row['Description']
```

```python
df['combined_features'] = df.apply(combined_features, axis = 1)
```

```python
df['combined_features'].head()
```

```python
df['combined_features'].iloc[1]
```

```python
cv = CountVectorizer()
```

```python
count_matrix = cv.fit_transform(df['combined_features'])
```

```python
count_matrix.toarray()
```

```python
count_matrix.shape
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
cosine_sim
```

```python
cosine_sim.shape
```

```python
df[df.CourseTitle == 'Google Cloud Big Data and Machine Learning Fundamentals']
```

```python
course  = "Google Cloud Big Data and Machine Learning Fundamentals"
```

```python
def get_id_from_course(title):
  return df[df.CourseTitle == course].index[0]
```

```python
course_index = get_id_from_course(course)
```

```python
course_index
```

```python
similar_course = list(enumerate(cosine_sim[course_index]))
```

```python
sorted_sim_courses = sorted(similar_course, key = lambda x:x[1], reverse = True)
```
